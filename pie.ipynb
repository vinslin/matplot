```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
label=["food","entertainment","travel","home"]
val=[3000,1344,2123,123]
```

```python
plt.pie(val,labels=label)
```

```python
#perfect circle
plt.axis("equal")
plt.pie(val,labels=label)
plt.show()#get rid of extra informations
```

```python
#perfect circle
plt.axis("equal")
plt.pie(val,labels=label,radius=2)#chart big panna
plt.show()#get rid of extra informations
```

```python
#perfect circle
plt.axis("equal")
plt.pie(val,labels=label,radius=1.5,autopct='%0.4f%%',explode=[0,0.2,0,0.5])#decimat points
plt.show()#get rid of extra informations
```

```python
#perfect circle
plt.axis("equal")
plt.pie(val,labels=label,radius=1,autopct='%0.4f%%',explode=[0,0.2,0,0.5],startangle=90)#decimat points
plt.legend(loc ="upper right")
plt.show()#get rid of extra informations
```

```python
#perfect circle
plt.axis("equal")
plt.pie(val,labels=label,radius=1.5,autopct='%0.4f%%',explode=[0,0.2,0,0.5], wedgeprops = {"edgecolor" : "black",
                      'linewidth': 2,
                      'antialiased': True})#decimat points
plt.show()#get rid of extra informations
```
