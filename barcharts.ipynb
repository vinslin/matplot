```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
company=["apple","orange","Raj","alibaba"]
revenue=[233,233,999,454]
```

```python
plt.bar(company,revenue)
```

```python
#another method 
x=np.arange(len(company))
x
```

```python
#x ticks have two arguments one is num value and the other one is name array
plt.xticks(x,company)
plt.xlabel("company")
plt.ylabel("revenue in billions")
plt.bar(x,revenue,width=0.3)
```

```python
profit=[23,23,500,45]
plt.xlabel("company")
plt.ylabel("revenue in billions")
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="profit")
plt.legend(loc='best')
```

```python
profit=[23,23,500,45]
plt.xticks(x,company)
plt.xlabel("company")
plt.ylabel("revenue in billions")
plt.bar(x-0.2,revenue,label="revenue",width=0.3)
plt.bar(x+0.2,profit,label="profit",width=0.3)
plt.legend(loc='best')
```

```python
#horizontal bar chart
plt.yticks(x,company)
plt.xlabel("company")
plt.ylabel("revenue in billions")
plt.barh(x-0.2,revenue,label="revenue")
plt.barh(x+0.2,profit,label="profit")
plt.legend(loc='best')
```

```python

# create data
df = pd.DataFrame([['A', 10, 20, 10, 26], ['B', 20, 25, 15, 21], ['C', 12, 15, 19, 6],['D', 10, 18, 11, 19]],columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])
# view data
print(df)

# plot data in stack manner of bar type
df.plot(x='Team', kind='bar', stacked=True,title='Stacked Bar Graph by dataframe')
plt.show()
```

```python
A = np.array([3,6,9,4,2,5])
X = np.arange(6)
 
# plot the bars
plt.bar(X, A, color = 'r')
plt.bar(X, -A, color = 'b')
plt.title("Back-to-Back Bar Chart")

plt.grid()
plt.show()
```
