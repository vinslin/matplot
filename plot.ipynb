```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[61,22,33,54,65,64,76]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("day")
plt.ylabel("temparature")
plt.title("temp")
plt.plot(x,y,color="green",linewidth=3,linestyle="dotted")
```

```python
#advanced
```

```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,alpha=0.5)
#alpha controls the visible of the line
```

```python
days=[1,2,3,4,5,6,7]
maxtemp=[61,22,33,54,65,64,76]
mintemp=[31,22,13,24,35,44,32]
avgtemp=[41,22,23,34,45,54,42]
```

```python
plt.plot(days,maxtemp,label="maxtemp",linestyle="dotted")
plt.plot(days,mintemp,label="mintemp")
plt.plot(days,avgtemp,label="avgtemp")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc="lower right")
plt.title("temperature")
```

```python
plt.plot(days,maxtemp,label="maxtemp",linestyle="dotted")
plt.plot(days,mintemp,label="mintemp")
plt.plot(days,avgtemp,label="avgtemp")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc="best")
plt.title("temperature")
plt.grid()
```

```python
plt.plot(days,maxtemp,label="maxtemp",linestyle="dotted")
plt.plot(days,mintemp,label="mintemp")
plt.plot(days,avgtemp,label="avgtemp")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc="best")
plt.title("temperature")
plt.fill_between(days,mintemp,avgtemp, color='green', alpha=0.5)
```

```python
plt.plot(days,maxtemp,label="maxtemp",linestyle="dotted")
plt.plot(days,mintemp,label="mintemp")
plt.plot(days,avgtemp,label="avgtemp")
plt.xlabel("days")
plt.ylabel("temp")
plt.legend(loc="best")
plt.title("temperature")
plt.fill_between(days,mintemp,avgtemp, color='green', alpha=0.5)
plt.fill_between(days,maxtemp,avgtemp, color='red', alpha=0.5)
```
