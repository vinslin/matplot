```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
bloodsugar_men=[101,77,65,75,87,72,67,167,200,267]
bloodsugar_women=[100,77,65,75,97,62,87,67,220,217]

plt.xlabel("sugar level")
plt.ylabel("how many")
plt.title("blood sugar analysis")
plt.hist([bloodsugar_men,bloodsugar_women],color=['green','red'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("sugar level")
plt.ylabel("how many")
plt.title("blood sugar analysis")
plt.hist([bloodsugar_men,bloodsugar_women],color=['green','red'],label=['men','women'],orientation="horizontal")
plt.legend()
```
